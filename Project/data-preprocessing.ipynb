```python
for file in tqdm.tqdm_notebook(glob(os.getcwd() + '/dataset/*.csv'), desc='1st loop'):
    data = pd.read_csv(file)
    data['datadate'] = pd.to_datetime(data['datadate'], format='%Y/%m/%d')
    
    
    for index, row in tqdm.tqdm_notebook(data.iterrows(), desc='2nd loop'):
        if row['tic'] in stockList:
            priceData.loc[row['datadate'], row['tic']] = row['prcod']
            sectorMap[row['tic']] = row['gsector']
            ratingMap[row['tic']] = row['spcsrc']
```

```python
dropCols = []
for col in priceData.columns:
    if priceData[col].dropna().index[0] > pd.to_datetime('2016-01-01'):
        print(col, priceData[col].dropna().index[0])
        dropCols.append(col)
```

```python
# choose price data after 2015-01-01
final = priceData.loc[priceData.index > '2016-01-01']
# remove all non rows
final.dropna(axis=0, how='all', inplace=True)
# remove stocks that listed after 2016-01-01
final.drop(dropCols, axis=1, inplace=True)

final.isnull().sum().sum()
```

```python
final_sectorMap = {}
final_ratingMap = {}
for k, v in sectorMap.items():
    if k in train.columns.values.tolist():
        final_sectorMap[k] = v
        final_ratingMap[k] = ratingMap[k]
```

```python
from scipy.stats import norm
```

```python
varDF = pd.DataFrame(index=train.columns, columns=[0.1, 0.05, 0.01])

for alpha in varDF.columns:
    varDF[alpha] = train.pct_change().dropna(axis=0, how='all').apply(
        lambda s: norm.ppf(alpha, s.mean(), s.std()), axis=0
    )
```
